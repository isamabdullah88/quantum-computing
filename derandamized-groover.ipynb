### Derandamized Groover Algorithm

```python
from qiskit import QuantumCircuit
from qiskit import quantum_info
from qiskit.primitives import StatevectorEstimator, StatevectorSampler
from qiskit.circuit import Parameter
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_distribution
from qiskit_ibm_runtime import QiskitRuntimeService, SamplerV2
from qiskit import transpile

import numpy as np
```

```python
circuit4 = QuantumCircuit(5, 3)

circuit4.h([0, 1, 2])
circuit4.ry(2*np.arcsin(np.sqrt(8)* np.sin(np.pi/ 10)), 3)
circuit4.x(4)
circuit4.h(4)
circuit4.barrier()

circuit4.x(1)
circuit4.mcx([0,1,2,3],4)
circuit4.x(1)
circuit4.barrier()
# circuit4.save_statevector()

circuit4.h([0, 1, 2])
circuit4.ry(-2*np.arcsin(np.sqrt(8)* np.sin(np.pi/ 10)), 3)
circuit4.x([0, 1, 2, 3])
circuit4.mcx([0,1,2,3],4)
circuit4.x([0, 1, 2, 3])
circuit4.h([0, 1, 2])
circuit4.ry(2*np.arcsin(np.sqrt(8)* np.sin(np.pi/ 10)), 3)
circuit4.barrier()


circuit4.x(1)
circuit4.mcx([0,1,2,3],4)
circuit4.x(1)
circuit4.barrier()

circuit4.h([0, 1, 2])
circuit4.ry(-2*np.arcsin(np.sqrt(8)* np.sin(np.pi/ 10)), 3)
circuit4.x([0, 1, 2, 3])
circuit4.mcx([0,1,2,3],4)
circuit4.x([0, 1, 2, 3])
circuit4.h([0, 1, 2])
circuit4.ry(2*np.arcsin(np.sqrt(8)* np.sin(np.pi/ 10)), 3)
circuit4.barrier()

# circuit4.x(2)
# circuit4.h(2)
# circuit4.ccx(0, 1, 2, ctrl_state="00")
# circuit4.h(2)
# circuit4.x(2)
# circuit4.barrier()

# circuit4.h([0, 1, 2])

circuit4.measure([0, 1, 2], [0, 1, 2])

circuit4.draw('mpl')



```

```python
sampler = StatevectorSampler()

results = sampler.run([(circuit4, None, 1000)]).result()

stats = results[0].data.c.get_counts()
plot_distribution(stats)
```
